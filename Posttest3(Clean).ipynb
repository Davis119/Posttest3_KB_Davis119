1. GRAFIK BATANG
Grafik data di bawah ini menunjukkan kolom 'Type' atau Krisis Kesehatan menurut Jenis Bencana, yaitu Bencana Sosial, Bencana Alam, dan Bencana Non Alam yang ada di dalam dataset tersebut.

```python
data_bar = df['Type'].value_counts()
plt.figure(figsize=(8, 8)) #untuk mengatur ukuran gambar dari grafik yang kita buat
plt.bar(x=data_bar.index, height=data_bar.values) #x=data_bar.index berfungsi utk membuat sumbu X
                                                  #height=data_bar.values berfungsi utk membuat sumbu Y 
plt.show() #menampilkan grafik
```

```python
#note ini menampilkan jumlah seluruh data

import pandas as pd
pd.options.display.max_columns = None
df = pd.read_csv('./Penyebab Kematian di Indonesia yang Dilaporkan - Raw.csv')
display(df)
```

```python
data_bar = df['Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.bar(x=data_bar.index, height=data_bar.values)
plt.xlabel('Jenis Data') #Memberi nama Sumbu X
plt.ylabel('Jumlah Data') #Memberi nama Sumbu Y

plt.title('Perbandingan Jumlah Data untuk Setiap Jenis') #Memberi Judul 
plt.show()
```

2.SCATTER PLOT
Dari data di bawah bisa disimpulkan bahwa sepanjang dari tahun 2000-2020 kematian tertinggi diakibatkan oleh
Bencana non Alam dan Penyakit.

```python
plt.scatter(df['Type'], df['Year']) #membuat scatter plot dimana sumbu X-nya adalah Jenis data dan 
                                    #sumbu Y-nya adalah Tahun
plt.title('Hubungan antara Jenis dan Tahun')
plt.xlabel('Jenis Data')
plt.ylabel('Tahun')
plt.show()
```

3.PIECHART
Dari data dibawah bisa Piechart ini memberikan kita gambaran visual tentang bagaimana setiap komponen(angka dalam piechart)
yaitu:
Cause, Type, Year, Data Redundancy, dan Total Deaths. Berkontribusi terhadap keselurahan data(100% data)

```python
import matplotlib.pyplot as plt

#buat data dan label
datapie = [1, 2, 3, 4, 5]
labels = ['Cause', 'Type', 'Year', 'Data Redundancy', 'Total Deaths']

# Ukuran gambar
plt.figure(figsize=(10, 10))

#membuat pie chart
plt.pie(datapie, labels=labels, autopct="%.2f%%")

#menampilkan pie chart
plt.title('Diagram Lingkaran Distribusi Data')
plt.show()
```

4.HEATMAP

Dari data di bawah ini bisa disimpulkan bahwa korelasi antar variabel2 memang mempengaruhi data secara signifikan

```python
#Memilih hanya kolom-kolom numerik(karena kalo bukan numerik maka codenya tidak bisa dijalankan)
df_numeric = df.select_dtypes(include=['number'])

correlation_matrix = df_numeric.corr() #menghitung matriks korelasi

#buat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Antara Variabel Numerik')
plt.show()
```
